```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
# loading the Boston Housing data set 
from sklearn.datasets import load_boston

boston = load_boston()
print(boston)
```

```python
# data = independent variables or the x values
# target = dependent variables or the y values
# column names also known as feature_names

data_frame_x = pd.DataFrame(boston.data, columns= boston.feature_names)
data_frame_y = pd.DataFrame(boston.target)
```

```python
print (data_frame_y)
```

```python
# getting some statistics from the dataset
data_frame_x.describe()
```

```python
# initializing the linear ligression model
reg = linear_model.LinearRegression()
```

```python
# splitting the data into testing and traing
x_train, x_test, y_train, y_test = train_test_split( data_frame_x, data_frame_y, test_size = 0.33, random_state = 42)

```

```python
# training the model with the training data
reg.fit(x_train, y_train)
```

```python
# printing coefficients/ weight of columns/features of the model
print(reg.coef_)
```

```python
# predictions of the test data
y_value_pred = reg.predict(x_test)
print(y_value_pred)
```

```python
y_vvalue_pred = reg.predict(x_train)
print(y_vvalue_pred)
```

```python
# printing the actual values
print(y_test)
```

```python
# check the model performance/accuracy (MSE -mean squared error)
print( np.mean( (y_value_pred - y_test)**2))
```

```python
# check the model performance/accuracy (MSE -mean squared error) and sklearn metrics
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_value_pred))
```
